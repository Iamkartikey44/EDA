```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Fifa.csv")
df.head()
```

```python
sns.jointplot(x='Age',y='Potential',
              joint_kws={'alpha':0.1,'s':3,'color':'yellow'},
              data=df,
             marginal_kws={"color":'green'})
plt.show();
```

<font color='red'>**Player Potential Decrease With Increase In Age**</font>

# <font color='blue'>**Sprint_speed vs Age**</font>

```python
sns.lmplot(x='Age',y='SprintSpeed',data=df,
          lowess=True,
          scatter_kws={'alpha':0.1,'color':'green','s':3},
          line_kws={'color':'brown'})
```

<font color='red'>**Sprint speed decreases with increase in age**</font>

# <font color='blue'>Which Foot Does Player Plays?</font>

```python
sns.lmplot(x='BallControl',y='Dribbling',data=df,
          col='Preferred Foot',
          scatter_kws={'alpha':0.1,'color':'green'},
          line_kws={'color':'cyan'})
```

# <font color='blue'>**Scores and Age With Foot**</font>

```python
plt.figure(figsize=(20,8))
sns.boxplot(x = 'Overall', y = 'Age', 
            hue = 'Preferred Foot',
            data = df,
            palette = 'rocket')
plt.show()
```
